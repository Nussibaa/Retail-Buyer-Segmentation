## 2. Preprocessing

This section handles **data cleaning, missing value imputation, and transformations** to prepare the dataset for modeling.  
We ensure the dataset is consistent, complete, and free of outliers or invalid categories.

### 2.1 Handling Missing Values

We check for null values and replace them with appropriate strategies:
- Annual income → median
- Number of teenagers → 0
- Education level → "Unknown"

```python
# check nulls
df.isna().sum()
```

```python
##check duplicate rows
## no duplicate rows
df.duplicated().sum()
```

```python
## replace nulls in annual income by the median 
median_income = df['annual_income'].median()
df['annual_income'] = df['annual_income'].fillna(median_income)
```

```python
## replace nulls in num_teenagers by 0
df['num_teenagers'] = df['num_teenagers'].fillna(0)
```

```python
## replace nulls in education_level by 'unknown'
df['education_level'] = df['education_level'].fillna('Unknown')
```

### 2.2 Standardizing Categories

We standardize categorical values to make them more understandable:
- Education Level → Primary, Secondary, Bachelor, Master, PhD, Unknown
- Marital Status → Merge "Alone" into "Single", "Together" into "Partnership", remove invalid categories ("Absurd", "YOLO")

```python
# Standardize education_level  
df['education_level'] = df['education_level'].replace({
    '2n Cycle': 'Secondary',
    'Basic': 'Primary',
    'Graduation': 'Bachelor',
    'PhD': 'PhD',
    'Master': 'Master',
    'unknown': 'Unknown'
})
```

```python
# Clean marital_status
df = df[~df['marital_status'].isin(['Absurd', 'YOLO'])] ## 4 rows dropped 
df['marital_status'] = df['marital_status'].replace({
    'Alone': 'Single',
    'Together': 'Partnership'
})
```

### 2.3 Removing Outliers

We calculate customer age from birth year and remove unrealistic ages (e.g., > 100 years).

```python
# Create Age column & Detect outlier
df['Age'] = 2025 - df['birth_year']
df['Age'].max()
```

```python
# Investigate how many rows are affected
df[df['Age'] > 100]
```

```python
df = df[df['Age'] <= 100]
```

```python
df['Age'].max()
```

```python
# Drop birth_year column
df = df.drop('birth_year', axis=1)
```

### Data Transformation & Visualization

```python
# Transform num_teenagers from float to int
df['num_teenagers'] = df['num_teenagers'].astype(int)
```

```python
# Transform date from object to datetime
df['signup_date'] = pd.to_datetime(df['signup_date'], dayfirst=True, infer_datetime_format=True)
```

```python
# Calculate number of days since sign-up
CURRENT_DATE = pd.to_datetime('2025-11-20')  # or pd.Timestamp.today()
df['days_since_signup'] = (CURRENT_DATE - df['signup_date']).dt.days
```

```python
# encode the categorical columns using on-hot encoding so the dataset is fully numerical
df_preprocessed = pd.get_dummies(df, columns=['education_level', 'marital_status'], drop_first=True)
df_preprocessed.to_csv('retail_customers_preprocessed.csv', index=False)
```

```python
numerical_cols = [
    'annual_income', 'num_children', 'num_teenagers', 'days_since_last_purchase',
    'spend_wine', 'spend_fruits', 'spend_meat', 'spend_fish', 'spend_sweets',
    'spend_gold', 'num_discount_purchases', 'num_web_purchases', 
    'num_catalog_purchases', 'num_store_purchases', 'web_visits_last_month',
    'days_since_signup', 'Age'
]
```

```python
# visualize numerical distributions 
# → to see which features are skewed and will need scaling/log-transformation 
#    before clustering (very important for K-means!)
df_preprocessed[numerical_cols].hist(figsize=(30,13))
plt.show()
```
